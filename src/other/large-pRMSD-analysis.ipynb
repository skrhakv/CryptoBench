```python
import pandas as pd

DATASET_PATH = '../../data/B-create-dataset/single-chain--p2rank-ligands/dataset_dataframe.csv'
LARGE_pRMSD_THRESHOLD = 6

cryptobench_df = pd.read_csv(DATASET_PATH)
large_prmsd = cryptobench_df[(
    cryptobench_df['apo_pocket_rms'] > LARGE_pRMSD_THRESHOLD)]
```

### Create scripts
let's create PYMOL visualizations for proteins with large pRMSD

```python
large_prmsd.head(2)
```

```python
large_prmsd.loc[large_prmsd.groupby('apo_structure')['apo_pocket_rms'].idxmax()]
```

```python
for row in large_prmsd.loc[large_prmsd.groupby('apo_structure')['apo_pocket_rms'].idxmax()].itertuples(): # take the largest pRMSD in the group to really show extremes

    apo_pdb_id = row.apo_structure
    holo_pdb_id = row.holo_structure
    apo_chain = row.apo_chains3
    holo_chain = row.holo_chains3
    apo_selection = row.apo_pocket_selection
    holo_selection = row.holo_pocket_selection
    pRMSD = row.apo_pocket_rms

    with open(f'large-pRMSD-pymol-scripts/{apo_pdb_id}-{holo_pdb_id}-{format(pRMSD, ".2f")}.pml', 'w') as pymol_script:

        pymol_script.write(f'reinitialize\n')

        pymol_script.write(f'fetch {apo_pdb_id}{apo_chain}\n')
        pymol_script.write(f'color blue, {apo_pdb_id}{apo_chain}\n')
        pymol_script.write(f'color aquamarine, {apo_selection}\n')
        pymol_script.write(f'show sticks, {apo_selection}\n')

        pymol_script.write(f'fetch {holo_pdb_id}{holo_chain}\n')
        pymol_script.write(f'color green, {holo_pdb_id}{holo_chain}\n')
        pymol_script.write(f'color yellow, {holo_selection}\n')
        pymol_script.write(f'show sticks, {holo_selection}\n')

        pymol_script.write(f'align {apo_pdb_id}{apo_chain}, {holo_pdb_id}{holo_chain} \n')
        pymol_script.write(f'zoom {apo_pdb_id}{apo_chain}\n')
```
